## Data Aggregation

```python
df_agg = df.groupby('model')['price'].agg(['count', 'mean', 'sum'])

display(df_agg)
```

```python
df_transmission = df.groupby('transmission')['price'].agg(['count', 'mean', 'sum'])

display(df_transmission)
```

```python
df_fuel = df.groupby('fuelType')['price'].agg(['count', 'mean', 'sum'])

display(df_fuel)
```

# 3. Data preprocessing before visualization

```python
# sort the dataframe based on the column 'count'
df_agg = df_agg.sort_values(by='count', ascending=False)

# create a new column percentage of the cars in each model
df_agg['percent_of_cars'] = (df_agg['count']/df_agg['count'].sum())*100

# convert the 'sum' column in millions
df_agg['sum'] = df_agg['sum']/1000000

# keep up to two decimal points
df_agg = df_agg.round(2)

# remove the 'count' column
df_agg = df_agg.drop('count', axis=1)

display(df_agg.head())
```

```python
# sort the dataframe based on the column 'count'
df_transmission = df_transmission.sort_values(by='count', ascending=False)

# create a new column percentage of the cars in each model
df_transmission['percent_of_cars'] = (df_transmission['count']/df_transmission['count'].sum())*100

# convert the 'sum' column in millions
df_transmission['sum'] = df_transmission['sum']/1000000

# keep up to two decimal points
df_transmission = df_transmission.round(2)

# remove the 'count' column
df_transmission = df_transmission.drop('count', axis=1)

display(df_transmission.head())
```

```python
# sort the dataframe based on the column 'count'
df_fuel = df_fuel.sort_values(by='count', ascending=False)

# create a new column percentage of the cars in each model
df_fuel['percent_of_cars'] = (df_fuel['count']/df_fuel['count'].sum())*100

# convert the 'sum' column in millions
df_fuel['sum'] = df_fuel['sum']/1000000

# keep up to two decimal points
df_fuel = df_fuel.round(2)

# remove the 'count' column
df_fuel = df_fuel.drop('count', axis=1)

display(df_fuel.head())
```

# 4. EDA using groupby/pivot_table and barplot (total 9 barplots), based on model, transmission, and fuelType

## Question: 
<br>A. What are the top 5 selling car models/transmission/fuelType in the dataset?
<br>B. What's the average selling price of the top 5 selling car models/transmission/fuelType?
<br>C. What's the total sale of the top 5 selling car models/transmission/fuelType?

```python
plt.figure(figsize=(25,20))

plt.subplot(3, 3, 1)
sns.barplot(x='percent_of_cars', y=df_agg.index, data=df_agg, orient='h', color='blue')
plt.title("Percentage of cars in each model")
plt.xlabel("Percentage of cars (%)")
# plt.grid()

plt.subplot(3, 3, 2)
sns.barplot(x='mean', y=df_agg.index, data=df_agg, orient='h', color='orange')
plt.title("Average resale value of the cars of each model")
plt.xlabel("Average price ($)")
plt.ylabel("")

plt.subplot(3, 3, 3)
sns.barplot(x='sum', y=df_agg.index, data=df_agg, orient='h', color='green')
plt.title("total resale value of the cars by each model")
plt.xlabel("Total price in million $")
plt.ylabel("")

plt.subplot(3, 3, 4)
sns.barplot(x='percent_of_cars', y=df_transmission.index, data=df_transmission, orient='h', color='blue')
plt.title("Percentage of cars in each transmission")
plt.xlabel("Percentage of cars (%)")
# plt.grid()

plt.subplot(3, 3, 5)
sns.barplot(x='mean', y=df_transmission.index, data=df_transmission, orient='h', color='orange')
plt.title("Average resale value of the cars of each transmission")
plt.xlabel("Average price ($)")
plt.ylabel("")

plt.subplot(3, 3, 6)
sns.barplot(x='sum', y=df_transmission.index, data=df_transmission, orient='h', color='green')
plt.title("total resale value of the cars by each transmission")
plt.xlabel("Total price in million $")
plt.ylabel("")

plt.subplot(3, 3, 7)
sns.barplot(x='percent_of_cars', y=df_fuel.index, data=df_fuel, orient='h', color='blue')
plt.title("Percentage of cars in each fuel")
plt.xlabel("Percentage of cars (%)")
# plt.grid()

plt.subplot(3, 3, 8)
sns.barplot(x='mean', y=df_fuel.index, data=df_fuel, orient='h', color='orange')
plt.title("Average resale value of the cars of each fuel")
plt.xlabel("Average price ($)")
plt.ylabel("")

plt.subplot(3, 3, 9)
sns.barplot(x='sum', y=df_fuel.index, data=df_fuel, orient='h', color='green')
plt.title("total resale value of the cars by each fuel")
plt.xlabel("Total price in million $")
plt.ylabel("")

plt.suptitle("Barplot of the Audi used car")
plt.tight_layout()
plt.show()
```
